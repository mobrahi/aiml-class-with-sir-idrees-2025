```python
# new_df['Price'].fillna(0 , inplace=True)

# we are using a central tendency to fill the nan values

new_df = df.copy()

mean_price = new_df['Price'].median()
print(mean_price)


new_df['Price'].fillna(mean_price, inplace=True)
new_df
```

# outliers in your data, instead of using mean, you can use median, can be a better representative of the central tendency #

```python
print(new_df['Price'].sort_values())
```

```python
# column fillna() # mean median mode[0] 


squared = new_df['Quantity'] ** 2 # square everything

new_quantity = squared / (squared ** 0.5) # divide the squares with their roots

new_quantity

new_df['Quantity'] = new_quantity
new_df

median_quantity = new_df['Quantity'].median()

new_df['Quantity'].fillna(median_quantity, inplace=True)
new_df
```

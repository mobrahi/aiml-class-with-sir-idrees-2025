```python
import sqlite3

import sqlite3 as sql
```

```python
connection = sql.connect('test.db')

cursor = connection.cursor()
```

```python
cursor.execute('''

CREATE TABLE students(
    id INT PRIMARY KEY,
    name VARCHAR(10),
    city VARCHAR(10),
    hobby VARCHAR(10)
)
               
''')
```

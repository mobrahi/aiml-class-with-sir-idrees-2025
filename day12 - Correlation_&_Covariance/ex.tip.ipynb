```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips') 
df.head()

print([df.columns])
```

```python
new_column = df['tip'] / df['total_bill'] * 100  #### broadcasting
df['tip_pct'] = new_column.round(2)

df
```

```python
correlation = df[['tip_pct', 'total_bill', 'size']].corr()
correlation
```

```python
sns.heatmap(correlation,annot=True)
```

```python
covariance = df[['tip_pct','total_bill','size']].cov()
covariance
```

```python
sns.heatmap(covariance, annot= True)
```

```python
df = sns.load_dataset('tips') 
df.head()
```

```python
sns.barplot(data=df,x="day",y="tip")
```

```python
sns.pairplot(data=df, hue="total_bill")
```

```python
sns.histplot(data=df, x="total_bill")
```

```python
df.describe().round(2)
```

```python
df[['total_bill','tip','size']].cov().round(2)

columns = ['total_bill','tip','size']
covariance = df[columns].cov()
covariance.round(2)
```

```python
columns = ['total_bill','tip','size']
correlation = df[columns].corr()
correlation.round(2)
```

```python
sns.heatmap(correlation, cmap='Greens',annot=True,fmt=".2f", vmin=-1,vmax=1)
```

```python
import plotly.express as px

barplot = px.bar(tips, x='day', y='total_bill')

barplot.show()
```

```python
mean_totalbill = tips['total_bill'].mean()
mean_totalbill
```

```python
#groupby any of the categorical columns
#aggregation

bill_by_day = tips.groupby('day', as_index=False)['total_bill'].count()
bill_by_day

#bill_by_sex = tips.groupby('sex')['total_bill'].max()
#bill_by_sex
```

```python
barplot = px.bar(bill_by_day, x= 'day', y= 'total_bill')

barplot.show()
```

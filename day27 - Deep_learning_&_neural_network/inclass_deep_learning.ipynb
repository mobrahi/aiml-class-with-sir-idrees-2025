```python
def neuron_output(x1, x2, w1, w2, b):
    
    z = w1 * x1 + w2 * x2 + b
    return z

x1 = 5
x2 = 2

w1 = 0.8
w2 = 1.2
b = -1.0

z = neuron_output(x1, x2, w1, w2, b)
print("Linear combination (z):", z)
    
```

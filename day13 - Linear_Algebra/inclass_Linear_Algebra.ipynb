```python
import numpy as np

x = np.array([[1,4,1],
              [2,5,2]])

w = np.array([10,10,10])

w2 = np.array([[10,2,3],
              [10,2,3],
              [10,2,3]])

mat_multiply_xw = np.dot(x,w)
mat_multiply_xw
```

```python
matmultiply_xw2 = np.dot(x,w2)
matmultiply_xw2
```

```python
matmultiply_xw2 = x @ w2
matmultiply_xw2
```

```python
matmultiply_xw2 = np.matmul(x,w2)
matmultiply_xw2
```

```python
import numpy as np

# Define a scalar
scalar = 3

# Define a NumPy array (vector)
vector = np.array([1, 2, 3, 4])

# Perform scalar multiplication on the vector
result_vector = scalar * vector
print("Result of scalar multiplication on vector:\n", result_vector)

# Define a NumPy array (matrix)
matrix = np.array([[1, 2],
                   [3, 4]])

# Perform scalar multiplication on the matrix
result_matrix = scalar * matrix
print("\nResult of scalar multiplication on matrix:\n", result_matrix)
```

```python
import pandas as pd

df = pd.DataFrame({'col1': [1, 2, 3], 'col2': [4, 5, 6]}, index=['A', 'B', 'C'])
print(df['A':'B']) # Slices rows with labels 'A' and 'B'
```

```python
import pandas as pd

df = pd.DataFrame({'col1': [1, 2, 3], 'col2': [4, 5, 6]}, index=['A', 'B', 'C'])

# Slicing rows by label
print(df.loc['A':'C'])

# Slicing rows and columns by label
print(df.loc['A':'B', 'col1':'col2'])
```

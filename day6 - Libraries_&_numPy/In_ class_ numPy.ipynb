```python
import numpy as np

# Create an array from 0 to 4 (stop=5, start=0, step=1 by default)
arr1 = np.arange(5)
print(f"Example 1: {arr1}")

# Create an array from 1 to 9 with a step of 2
arr2 = np.arange(1, 10, 2)
print(f"Example 2: {arr2}")

# Create an array with floating-point numbers
arr3 = np.arange(0.5, 2.5, 0.5)
print(f"Example 3: {arr3}")

# Create a descending array using a negative step
arr4 = np.arange(10, 0, -2)
print(f"Example 4: {arr4}")
```

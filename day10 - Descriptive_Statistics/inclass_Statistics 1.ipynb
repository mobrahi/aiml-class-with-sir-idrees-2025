```python
import numpy as np

class1 = np.array([50,47,51,49,47])
class2 = np.array([5,20,50,80,130])

mean = sum(class1)/len(class1)
print("mean is: ",mean)

variance1 = sum((class1 - mean)**2)/len(class1)
print(variance1) 
```

```python
mean = sum(class2)/len(class2)
print("mean is: ",mean)

sum_of_differences = np.array([((class2 - mean)**2)])
variance2 = (sum_of_differences.sum())/len(class2)
print(variance2) 
```

```python
std1 = variance1**0.5
print(std1)
std2 = variance2**0.5
print(std2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("/Users/mf/ai_ml_class/day32 - mini_project/rainfall.csv")


# 3. Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[['Mean temperature-Min (oC)', 'Mean temperature-Max (oC)', 
                'Rainfall-Total (mm)', 'Rainfall-No. of days', 
                'Mean relative humidity (%)', 
                'Height above mean sea level in metres']].corr(), 
            annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

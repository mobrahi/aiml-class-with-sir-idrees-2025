```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report 

df = pd.read_csv('/Users/mf/ai_ml_class/day32 - project_finale/_MalaysiaFloodDataset_MalaysiaFloodDataset.csv')
```

```python
df = df.rename(columns={'0V': 'NOV'})
```

```python
month_cols = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
df_long = df.melt(id_vars=['STATE', 'DISTRICT', 'YEAR', 'FLOOD '], 
                  value_vars=month_cols, 
                  var_name='Month_Name', 
                  value_name='Rainfall')
```

```python
month_map = {m: i+1 for i, m in enumerate(month_cols)}
df_long['Month'] = df_long['Month_Name'].map(month_map)
df_long['Date'] = pd.to_datetime(df_long[['YEAR', 'Month']].assign(DAY=1))
df_long = df_long.sort_values(['DISTRICT', 'Date'])
```

```python
df_kelantan = df_long[df_long['STATE'] == 103].pivot(index='Date', columns='DISTRICT', values='Rainfall')
```

```python
df_kelantan['KB_Rain'] = df_kelantan[103001]
df_kelantan['KK_Rain_Lag1'] = df_kelantan[103003].shift(1)
```

```python
df_final = df_kelantan[['KB_Rain', 'KK_Rain_Lag1']].dropna()
```

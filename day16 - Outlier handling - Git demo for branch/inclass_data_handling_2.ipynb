```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
data = sns.load_dataset('tips') 

df = pd.DataFrame(data)
df['total_bill'].head()
```

```python
plt.boxplot(df['total_bill']);
```

```python
z_scores = stats.zscore(df['total_bill'])
```

```python
z_scores_total_bill = stats.zscore(df['total_bill'])
z_scores_total_bill
```

```python
threshold = 2

df2 = df[ (-threshold < z_scores_total_bill) & (z_scores_total_bill < threshold)] 
df2.head()
```

```python
plt.boxplot(df2['total_bill']);
```

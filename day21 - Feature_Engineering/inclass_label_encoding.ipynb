# FEATURE ENGINEERING

it is the process of transforming raw data into features that make machine learning algorithms work better.
-to improve performance of ML models
-it can improve quality and relevance of features

Categorical Features

    -Groups of data
        -will include Binary Features (0 and 1), Survived - Yes NO, Bool data type
        -Category data type - use for columns with limited, fixed number of unique values (low cardinality) ie gender, education level, or product size
        -Ordinal data type - you can set up a scale in the categories - Small Medium Large - Cold warm hot, small height medium height large height

Numerical Features

    -where there are numbers
    -will include float and integers both
    -continuos values,
    -grouping up will not make much sense (unless doing clustering)

- Encoding

```python
# label encoding
# column that u encode
# manually .map function

penguins['species'].unique()

species_codes = {
    'Adelie': 0,
    'Chinstrap': 1,
    'Gentoo': 2
}

penguins['species_encoded'] = penguins['species'].map(species_codes)
penguins.sample(10)
```

```python
penguins['island'].unique()

island_codes = {
    'Torgersen':0, 
    'Biscoe':1, 
    'Dream':2
}

penguins['island_encoded'] = penguins['island'].map(island_codes)
penguins.sample(10)
```

```python
x, y = pd.factorize(penguins['island'])

penguins ['island_factorized'] = x
penguins
```

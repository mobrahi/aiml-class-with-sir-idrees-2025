```python

iris_one_hot = iris.copy()

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

categorical_cols = ['species']

ecoded_values = encoder.fit_transform(iris_one_hot[categorical_cols]) # species

#name of new columns, that don't exist yet
new_columns = encoder.get_feature_names_out(categorical_cols)
new_columns
# ['species_setosa', 'species_versicolor', 'species_virginica']

#iris_one_hot

#adding the columns to the dataframe

iris_one_hot[new_columns] = ecoded_values

iris_one_hot.drop(columns=['species'], inplace=True)
iris_one_hot
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = pd.DataFrame({
    'age': [20,50,100],
    'income': [20000,50000,100000]
})

scaler = MinMaxScaler()

normalize_data = scaler.fit_transform(data)

df_normalized = pd.DataFrame(normalize_data, columns=['age', 'income'])
df_normalized
```

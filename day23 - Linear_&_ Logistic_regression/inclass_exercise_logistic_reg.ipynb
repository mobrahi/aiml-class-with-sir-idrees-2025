```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
titanic = sns.load_dataset('titanic')

titanic.head(2)
```

```python
label_codes, label_names = pd.factorize(titanic['sex'])
titanic ['sex_factorized'] = label_codes

X = titanic[['survived']]
y = titanic['sex_factorized']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state= 42, test_size=0.2)
```

```python
titanic.isna().sum()
```

```python
titanic.info()
```

```python
lr = LogisticRegression() # classification

lr.fit(X_train, y_train) # fitting here, must be numeric

y_pred = lr.predict(X_test)

lr.score(X_test, y_test)

print(f"Accuracy Score: {lr.score(X_test, y_test):.4f}")
```

```python
r2 = r2_score(y_test, y_pred)  
mae = mean_absolute_error(y_test,y_pred)

print(r2_score(y_test, y_pred), mean_absolute_error(y_test, y_pred))
```
